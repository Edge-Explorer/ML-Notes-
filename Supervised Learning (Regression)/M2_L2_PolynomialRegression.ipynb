```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(42)

X=6*np.random.rand(100, 1)-3 # range: [-3, 3]
y=0.5*X**2+X+2+np.random.randn(100, 1)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X, y)
y_pred_lin=lin_reg.predict(X)
```

```python
poly_features=PolynomialFeatures(degree=2, include_bias=False)
X_poly=poly_features.fit_transform(X)

poly_reg=LinearRegression()
poly_reg.fit(X_poly, y)
y_pred_poly=poly_reg.predict(X_poly)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="blue", label="Data points", alpha=0.6)
plt.scatter(X, y_pred_lin, color="red", label="Linear Regression", alpha=0.6)
plt.scatter(X, y_pred_poly, color="green", label="Polynomial Regression (degree=2)", alpha=0.6)
plt.title("Linear vs Polynomial Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
print("Linear Regression MSE:", mean_squared_error(y, y_pred_lin))
print("Polynomial Regression MSE:", mean_squared_error(y, y_pred_poly))
print("Linear Regression R:", r2_score(y, y_pred_lin))
print("Polynomial Regression R:", r2_score(y, y_pred_poly))
```
