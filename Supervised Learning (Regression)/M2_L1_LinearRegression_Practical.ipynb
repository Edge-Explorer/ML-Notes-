```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(42)

X=2*np.random.rand(100, 1) #So now X contains 100 random values between 0 and 2.You can think of it like we are generating 100 random “feature” values (X) for our small dataset — something that could represent, say, years of experience or square footage.
y=4+3*X+np.random.randn(100, 1)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train, y_train)

print("Learned Slope (m):", model.coef_[0][0])
print("Learned Intercept (b):", model.intercept_[0])
```

```python
y_pred=model.predict(X_test)
```

```python
mse=mean_squared_error(y_test, y_pred) #MSE is not a percentage.
r2=r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R2 Score (r2):", r2)
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```
