```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

```python
# A small sample dataset
data={
    'Area_sqft':[1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500],
    'Bedrooms':[2, 3, 3, 4, 4, 5, 5, 6],
    'Location_score':[6, 7, 7, 8, 8, 9, 9, 10],
    'Price_Lakhs':[50, 65, 80, 92, 110, 125, 140, 155]
}

df=pd.DataFrame(data)
df
```

### Separate Features (X) and Labels (y)

```python
X=df[['Area_sqft', 'Bedrooms', 'Location_score']] # Feature matrix
y=df['Price_Lakhs'] # Label vector

print("Feature Matrix (X):")
print(X.head())

print("\nLabel Vector (y):")
print(y.head())
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=42)

print("Training set size:", X_train.shape[0])
print("Test set size:", X_test.shape[0])
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Area_sqft'], df['Price_Lakhs'], color='blue', label='All Data')
plt.xlabel('Area (sqft)')
plt.ylabel('Price(Lakhs)')
plt.title('Area vs Price')
plt.legend()
plt.show()
```
