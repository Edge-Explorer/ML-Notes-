```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 
```

```python
np.random.seed(42)

X=np.random.rand(50,1)*10 #50 random X values between o and 10
y=3*X+4+np.random.randn(50,1)*2 #Linear relation + nosie 

plt.scatter(X, y, color='blue')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Sample Data (y=3x+4+noise)")
plt.show()
```

```python
model=LinearRegression()
model.fit(X, y)

print("Learned slope (m):", model.coef_[0][0])
print("Learned intercept (b):", model.intercept_[0])
```

```python
y_pred=model.predict(X)

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Learned line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Model Learning: Fitting a line")
plt.legend()
plt.show()
```

```python
x_new= np.array([[7]])
y_new=model.predict(x_new)
print(f"When X=7, predicted y={y_new[0][0]:.2f}")
```
