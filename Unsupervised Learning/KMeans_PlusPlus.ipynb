```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X, y=make_blobs(
    n_samples=500,
    centers=4,
    cluster_std=2,
    random_state=42
)
plt.scatter(X[:,0], X[:,1])
plt.title("Raw Data (Unscaled)")
plt.show()
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

plt.scatter(X_scaled[:,0], X_scaled[:,1])
plt.title("Scaled Data")
plt.show()
```

```python
kmeans_pp=KMeans(
    n_clusters=3,
    init="k-means++",
    n_init=10,
    random_state=42
)
kmeans_pp.fit(X_scaled)
labels_pp=kmeans_pp.labels_
centers_pp=kmeans_pp.cluster_centers_
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels_pp)
plt.scatter(centers_pp[:,0], centers_pp[:,1],
           marker="X", s=200)
plt.title("K-Means++ Initialization")
plt.show()
print("WCSS:", kmeans_pp.inertia_)
```
