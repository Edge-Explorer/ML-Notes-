```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler 
```

```python
X, y=make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=1.2,
    random_state=42
)

plt.scatter(X[:,0], X[:,1])
plt.title("Raw Data (Unscaled)")
plt.show()
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

plt.scatter(X_scaled[:,0], X_scaled[:,1])
plt.title("Scaled Data")
plt.show()
```

```python
kmeans_random=KMeans(
    n_clusters=3,
    init="random",
    n_init=1,
    random_state=69
)

kmeans_random.fit(X_scaled)
labels_random=kmeans_random.labels_
centers_random=kmeans_random.cluster_centers_
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels_random)
plt.scatter(centers_random[:,0], centers_random[:,1],
           marker='X', s=200)

plt.title("K-Means (Random Initialization)")
plt.show()
print("WCSS:", kmeans_random.inertia_)
```
